Problem: Predict whether a customer will churn(Yes/No)
Type : Binary Classification
Target Colomn: Churn


```python
import pandas as pd



df= pd.read_csv('telco_churn.csv')
```

```python
df.head()
```

```python
df.drop(columns=['customerID'], inplace=True)
df.head()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors = 'coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['TotalCharges']= df['TotalCharges'].fillna(0).astype('float64')  
```

Binary classification → 0 / 1


```python
df["Churn"] = df["Churn"].map({'Yes': 1, 'No': 0})
```

One-Hot Encoding

```python
df.shape
```

```python
df_encoding = pd.get_dummies(df, drop_first = True)
```

```python
df_encoding.shape
df_encoding
```

Train-Test-Split


x(Feature) is independent variable i.e all column except Churn.  
y(Target) column we want to predict.  


To avoid data leakage we dropped Churn from x.    

```python
from sklearn.model_selection import train_test_split
x = df_encoding.drop('Churn', axis =1)
y = df_encoding['Churn']
```

```python
#Splitting the data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42, stratify = y)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
y_train.value_counts(normalize=True)
y_test.value_counts(normalize=True)
```

```python
x_train.head()
y_train.head()

```

Build Your FIRST Model  

Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = LogisticRegression(max_iter=5000)
```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print(y_pred)
```

```python
y_prob = model.predict_proba(x_test)[:, 1]
```

```python
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```
